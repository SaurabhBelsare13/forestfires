```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline

```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
# scatter plot
plt.scatter(df['Height'], df['Weight'])
plt.xlabel('Height')
plt.ylabel('Weight')
```

```python
# divide features into dependent and independent
# Always make sure that our independent feature is in the form of dataframe or 2d array(row,col)
X = df[['Height']]
# dependent feature can be in series or 1-d array
y = df['Weight']
```

```python
X
```

```python
y
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25,random_state = 20)
```

```python
X_train.shape
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
# Applying Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
# Coefficient or slop : one unit movement in x - axis what is the unit change with respect to y axis
# Intercept : what is the value x(Height) is 0, what is the value of y(Weight).
print('Coefficient or slope: ' , regression.coef_)
print('Intercept: ', regression.intercetp_)
```
